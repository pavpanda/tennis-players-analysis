```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("serena_data.csv")
df
```

```python
def get_row(df, row_name):
    return df.loc[df['Category'] == row_name]
def plot_row(df, row_name, plot_line, plot_connections, starting_year, ending_year, title, xlabel, ylabel, file, onlyIntegers, boundsArray, intervals):    
    y = np.array(get_row(df, row_name)).flatten()[1:]
    x = np.array(range(starting_year, ending_year + 1))
    
    print(len(x))
    print(len(y))

    if (plot_connections):
        plt.scatter(x, y)
        plt.plot(x, y)
    else:
        plt.plot(x, y, 'o')
    if plot_line:
        plt.plot(x, m*x + b)
        print(m)
        print(b)
        
    if (onlyIntegers):
        plt.yticks(range(boundsArray[0], boundsArray[1] + 1))
    plt.xticks(range(starting_year, ending_year + 1, intervals))

    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.savefig(file)
```

```python
plot_row(df, "Grand slams", False, True, 1995, 2020, "Number of Grand Slam titles vs Time", "Time (years)", "Number of Grand Slam titles", "serenaslams.png", True, [0,4], 3)
```
