```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("federer_data.csv")
df
```

```python
def get_row(df, row_name):
    return df.loc[df['Category'] == row_name]
def plot_row(df, row_name, plot_line, plot_connections, starting_year, ending_year):    
    y = np.array(get_row(df, row_name)).flatten()[1:]
    x = np.array(range(starting_year, ending_year + 1))
    
    print(len(x))
    print(len(y))

    if (plot_connections):
        plt.scatter(x, y)
        plt.plot(x, y)
    else:
        plt.plot(x, y, 'o')
    if plot_line:
        plt.plot(x, m*x + b)
        print(m)
        print(b)

    plt.title("Number of Grand Slam titles vs Time")
    plt.xlabel('Time (Year)')
    plt.ylabel('Number of Grand Slam titles')
    plt.savefig("federerslams.png")
```
